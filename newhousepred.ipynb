```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

# Set plot style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("All libraries imported successfully!")
```

```python
df_train = pd.read_csv('df_train.csv')
df_test = pd.read_csv('df_test (2).csv')

print("Dataset Information:")
print("="*70)
print(f"Train shape: {df_train.shape}")
print(f"Test shape: {df_test.shape}")
print(f"\n Datasets loaded successfully!")
```

```python
print("\n" + "="*70)
print("EXPLORATORY DATA ANALYSIS")
print("="*70)

# Display first few rows
print("\n First 5 rows of training data:")
print(df_train.head())

# Check data types
print("\n Data Types:")
print(df_train.dtypes)

# Statistical summary
print("\n Statistical Summary:")
print(df_train.describe())

# Check for missing values
print("\n Missing Values:")
print(f"Train missing values: {df_train.isnull().sum().sum()}")
print(f"Test missing values: {df_test.isnull().sum().sum()}")

# Price distribution
print("\nPrice Statistics:")
print(f"Mean Price: ${df_train['price'].mean():,.2f}")
print(f"Median Price: ${df_train['price'].median():,.2f}")
print(f"Min Price: ${df_train['price'].min():,.2f}")
print(f"Max Price: ${df_train['price'].max():,.2f}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Price distribution
axes[0, 0].hist(df_train['price'], bins=50, color='skyblue', edgecolor='black')
axes[0, 0].set_xlabel('Price')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Price Distribution')

# Price vs Living Area
axes[0, 1].scatter(df_train['living_in_m2'], df_train['price'], alpha=0.5, color='coral')
axes[0, 1].set_xlabel('Living Area (m)')
axes[0, 1].set_ylabel('Price')
axes[0, 1].set_title('Price vs Living Area')

# Price vs Grade
axes[1, 0].boxplot([df_train[df_train['grade']==i]['price'] for i in sorted(df_train['grade'].unique())])
axes[1, 0].set_xlabel('Grade')
axes[1, 0].set_ylabel('Price')
axes[1, 0].set_title('Price Distribution by Grade')

# Price vs Bedrooms
axes[1, 1].boxplot([df_train[df_train['bedrooms']==i]['price'] for i in sorted(df_train['bedrooms'].unique())])
axes[1, 1].set_xlabel('Bedrooms')
axes[1, 1].set_ylabel('Price')
axes[1, 1].set_title('Price Distribution by Bedrooms')

plt.tight_layout()
plt.show()
```

```python
print("\n" + "="*70)
print("FEATURE CORRELATION ANALYSIS")
print("="*70)

# Prepare data for correlation (convert boolean to int)
df_corr = df_train.copy()
bool_cols = df_corr.select_dtypes(include='bool').columns
df_corr[bool_cols] = df_corr[bool_cols].astype(int)
df_corr = df_corr.drop(['date'], axis=1)

# Calculate correlation with price
correlation_with_price = df_corr.corr()['price'].sort_values(ascending=False)
print("\n Correlation with Price:")
print(correlation_with_price)

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_corr.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# CELL 6: Data Preprocessing & Feature Engineering
# ============================================================================
print("\n" + "="*70)
print("DATA PREPROCESSING & FEATURE ENGINEERING")
print("="*70)

def preprocess_data(df, is_train=True):
    """
    Preprocess the dataset with feature engineering
    """
    df = df.copy()

    # Drop date column (not useful for prediction)
    df = df.drop(['date'], axis=1)

    # Convert boolean columns to integer
    bool_cols = df.select_dtypes(include='bool').columns
    df[bool_cols] = df[bool_cols].astype(int)

    # Feature Engineering: Create interaction features
    df['price_per_sqm'] = df['price'] / df['living_in_m2'] if 'price' in df.columns else 0
    df['bathroom_bedroom_ratio'] = df['real_bathrooms'] / (df['bedrooms'] + 1)
    df['total_rooms'] = df['bedrooms'] + df['real_bathrooms']
    df['luxury_score'] = (df['grade'] * df['living_in_m2']) / 100
    df['quality_index'] = df['grade'] * (df['perfect_condition'] + 1)

    return df

# Preprocess training data
df_train_processed = preprocess_data(df_train, is_train=True)
df_test_processed = preprocess_data(df_test, is_train=True)

print("Data preprocessing completed!")
print(f"New shape after feature engineering: {df_train_processed.shape}")
print(f"\nNew features created:")
print("- price_per_sqm: Price per square meter")
print("- bathroom_bedroom_ratio: Bathrooms to bedrooms ratio")
print("- total_rooms: Total number of rooms")
print("- luxury_score: Grade  Living area score")
print("- quality_index: Grade  Condition score")
```

```python
print("\n" + "="*70)
print("PREPARING FEATURES AND TARGET VARIABLES")
print("="*70)

# Separate features and target
X_train = df_train_processed.drop(['price', 'price_per_sqm'], axis=1)
y_train = df_train_processed['price']

X_test = df_test_processed.drop(['price', 'price_per_sqm'], axis=1)
y_test = df_test_processed['price']

print(f" Training features shape: {X_train.shape}")
print(f" Training target shape: {y_train.shape}")
print(f" Testing features shape: {X_test.shape}")
print(f" Testing target shape: {y_test.shape}")

print(f"\nFeatures used for training:")
print(list(X_train.columns))
```

```python
print("\n" + "="*70)
print("FEATURE SCALING")
print("="*70)

# Use RobustScaler (better for outliers)
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(" Features scaled using RobustScaler")
print("   (Better than StandardScaler for handling outliers)")
```

```python
print("\n" + "="*70)
print("TRAINING MULTIPLE REGRESSION MODELS")
print("="*70)

# Dictionary to store models and their performance
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=10),
    'Lasso Regression': Lasso(alpha=10),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")

    # Train the model
    model.fit(X_train_scaled, y_train)

    # Make predictions
    y_pred = model.predict(X_test_scaled)

    # Calculate metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    # Store results
    results[name] = {
        'model': model,
        'r2': r2,
        'mae': mae,
        'rmse': rmse,
        'mape': mape,
        'predictions': y_pred
    }

    print(f"    R Score: {r2:.4f}")
    print(f"    MAE: ${mae:,.2f}")
    print(f"    RMSE: ${rmse:,.2f}")
    print(f"    MAPE: {mape:.2f}%")
```

```python
print("\n" + "="*70)
print("MODEL PERFORMANCE COMPARISON")
print("="*70)

# Create comparison dataframe
comparison_df = pd.DataFrame({
    'Model': list(results.keys()),
    'R Score': [results[m]['r2'] for m in results.keys()],
    'MAE ($)': [results[m]['mae'] for m in results.keys()],
    'RMSE ($)': [results[m]['rmse'] for m in results.keys()],
    'MAPE (%)': [results[m]['mape'] for m in results.keys()]
}).sort_values('R Score', ascending=False)

print("\n", comparison_df.to_string(index=False))

# Find best model
best_model_name = comparison_df.iloc[0]['Model']
best_model = results[best_model_name]['model']

print(f"\n BEST MODEL: {best_model_name}")
print(f"   R Score: {results[best_model_name]['r2']:.4f}")
print(f"   MAE: ${results[best_model_name]['mae']:,.2f}")
print(f"   RMSE: ${results[best_model_name]['rmse']:,.2f}")

# Visualize model comparison
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# R Score comparison
axes[0].barh(comparison_df['Model'], comparison_df['R Score'], color='steelblue')
axes[0].set_xlabel('R Score')
axes[0].set_title('Model R Score Comparison')
axes[0].set_xlim(0, 1)

# MAE comparison
axes[1].barh(comparison_df['Model'], comparison_df['MAE ($)'], color='coral')
axes[1].set_xlabel('Mean Absolute Error ($)')
axes[1].set_title('Model MAE Comparison')

plt.tight_layout()
plt.show()
```

```python
# ============================================================================
# CELL 11: Hyperparameter Tuning for Best Model
# ============================================================================
print("\n" + "="*70)
print("HYPERPARAMETER TUNING")
print("="*70)

# Make sure best_model is properly assigned
best_model = results[best_model_name]['model']

if best_model_name == 'Random Forest':
    print("\nTuning Random Forest hyperparameters...")

    param_grid = {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, 30, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    grid_search = GridSearchCV(
        RandomForestRegressor(random_state=42, n_jobs=-1),
        param_grid,
        cv=3,
        scoring='r2',
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train_scaled, y_train)

    print(f"\nBest parameters: {grid_search.best_params_}")
    print(f"Best CV R Score: {grid_search.best_score_:.4f}")

    # Use tuned model
    best_model = grid_search.best_estimator_

elif best_model_name == 'Gradient Boosting':
    print("\n Tuning Gradient Boosting hyperparameters...")

    param_grid = {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 5, 7],
        'min_samples_split': [2, 5, 10]
    }

    grid_search = GridSearchCV(
        GradientBoostingRegressor(random_state=42),
        param_grid,
        cv=3,
        scoring='r2',
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train_scaled, y_train)

    print(f"\nBest parameters: {grid_search.best_params_}")
    print(f" Best CV R Score: {grid_search.best_score_:.4f}")

    # Use tuned model
    best_model = grid_search.best_estimator_
else:
    print(f"\n Using {best_model_name} without hyperparameter tuning")
    print(f"   (Tuning only implemented for Random Forest and Gradient Boosting)")

print(f"\n Final model ready: {best_model_name}")
```

```python
# CELL 12: Final Model Evaluation
# ============================================================================
print("\n" + "="*70)
print(" FINAL MODEL EVALUATION")
print("="*70)

# Debug: Check if best_model exists
print(f"Using model: {type(best_model).__name__}")
print(f"Model object: {best_model}\n")

# Make final predictions
y_pred_final = best_model.predict(X_test_scaled)

# Verify predictions were created
print(f" Predictions created: {len(y_pred_final)} values")
print(f" Predictions shape: {y_pred_final.shape}")

# Calculate final metrics
final_r2 = r2_score(y_test, y_pred_final)
final_mae = mean_absolute_error(y_test, y_pred_final)
final_rmse = np.sqrt(mean_squared_error(y_test, y_pred_final))
final_mape = np.mean(np.abs((y_test - y_pred_final) / y_test)) * 100

print(f"\nFinal Model Performance:")
print(f"   Model: {best_model_name}")
print(f"   R Score: {final_r2:.4f} ({final_r2*100:.2f}%)")
print(f"   MAE: ${final_mae:,.2f}")
print(f"   RMSE: ${final_rmse:,.2f}")
print(f"   MAPE: {final_mape:.2f}%")

# Interpretation
print(f"\n Model Interpretation:")
print(f"    The model explains {final_r2*100:.2f}% of price variance")
print(f"    Average prediction error: ${final_mae:,.2f}")
print(f"    Predictions are off by {final_mape:.2f}% on average")
```

```python
# CELL 13: Visualize Predictions
# ============================================================================
print("\n" + "="*70)
print(" PREDICTION VISUALIZATIONS")
print("="*70)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Actual vs Predicted
axes[0, 0].scatter(y_test, y_pred_final, alpha=0.5, color='blue')
axes[0, 0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0, 0].set_xlabel('Actual Price')
axes[0, 0].set_ylabel('Predicted Price')
axes[0, 0].set_title('Actual vs Predicted Prices')

# 2. Residual Plot
residuals = y_test - y_pred_final
axes[0, 1].scatter(y_pred_final, residuals, alpha=0.5, color='green')
axes[0, 1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[0, 1].set_xlabel('Predicted Price')
axes[0, 1].set_ylabel('Residuals')
axes[0, 1].set_title('Residual Plot')

# 3. Distribution of Errors
axes[1, 0].hist(residuals, bins=50, color='orange', edgecolor='black')
axes[1, 0].set_xlabel('Prediction Error')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('Distribution of Prediction Errors')

# 4. Error Percentage Distribution
error_pct = (np.abs(residuals) / y_test) * 100
axes[1, 1].hist(error_pct, bins=50, color='purple', edgecolor='black')
axes[1, 1].set_xlabel('Error Percentage (%)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('Distribution of Error Percentage')

plt.tight_layout()
plt.show()
```

```python
# ============================================================================
# CELL 15: Test Model with Sample Predictions
# ============================================================================
print("\n" + "="*70)
print("TESTING MODEL WITH SAMPLE PREDICTIONS")
print("="*70)

# Select 10 random samples from test set
sample_indices = np.random.choice(len(X_test), 10, replace=False)
X_sample = X_test.iloc[sample_indices]
y_sample_actual = y_test.iloc[sample_indices]

# Scale and predict
X_sample_scaled = scaler.transform(X_sample)
y_sample_pred = best_model.predict(X_sample_scaled)

# Create comparison dataframe
sample_results = pd.DataFrame({
    'Actual Price': y_sample_actual.values,
    'Predicted Price': y_sample_pred,
    'Difference': y_sample_actual.values - y_sample_pred,
    'Error %': np.abs((y_sample_actual.values - y_sample_pred) / y_sample_actual.values) * 100
})

print("\n Sample Predictions:")
print(sample_results.to_string(index=False))

print(f"\nAverage Error: ${np.abs(sample_results['Difference']).mean():,.2f}")
print(f" Average Error %: {sample_results['Error %'].mean():.2f}%")
```

```python
# ============================================================================
# CELL 16: Create Prediction Function for New Data
# ============================================================================
print("\n" + "="*70)
print("CREATE PREDICTION FUNCTION")
print("="*70)

def predict_house_price(bedrooms, grade, has_basement, living_in_m2, renovated,
                        nice_view, perfect_condition, real_bathrooms, has_lavatory,
                        single_floor, month, quartile_zone):
    """
    Predict house price based on input features

    Parameters:
    -----------
    All the features from the dataset

    Returns:
    --------
    Predicted house price
    """
    # Create input dataframe
    input_data = pd.DataFrame({
        'bedrooms': [bedrooms],
        'grade': [grade],
        'has_basement': [int(has_basement)],
        'living_in_m2': [living_in_m2],
        'renovated': [int(renovated)],
        'nice_view': [int(nice_view)],
        'perfect_condition': [int(perfect_condition)],
        'real_bathrooms': [real_bathrooms],
        'has_lavatory': [int(has_lavatory)],
        'single_floor': [int(single_floor)],
        'month': [month],
        'quartile_zone': [quartile_zone]
    })

    # Feature engineering (same as training)
    input_data['bathroom_bedroom_ratio'] = input_data['real_bathrooms'] / (input_data['bedrooms'] + 1)
    input_data['total_rooms'] = input_data['bedrooms'] + input_data['real_bathrooms']
    input_data['luxury_score'] = (input_data['grade'] * input_data['living_in_m2']) / 100
    input_data['quality_index'] = input_data['grade'] * (input_data['perfect_condition'] + 1)

    # Scale features
    input_scaled = scaler.transform(input_data)

    # Predict
    prediction = best_model.predict(input_scaled)[0]

    return prediction

# Test the function
print("\n Testing prediction function with example:")
print("\nInput:")
print("  - Bedrooms: 3")
print("  - Grade: 3")
print("  - Has Basement: Yes")
print("  - Living Area: 200 m")
print("  - Renovated: No")
print("  - Nice View: Yes")
print("  - Perfect Condition: Yes")
print("  - Bathrooms: 2")
print("  - Has Lavatory: Yes")
print("  - Single Floor: No")
print("  - Month: 6")
print("  - Quartile Zone: 3")

test_price = predict_house_price(
    bedrooms=3,
    grade=2,
    has_basement=True,
    living_in_m2=342,
    renovated=False,
    nice_view=False,
    perfect_condition=False,
    real_bathrooms=3,
    has_lavatory=False,
    single_floor=False,
    month=1,
    quartile_zone=4
)

print(f"\n Predicted Price: ${test_price:,.2f}")
```
